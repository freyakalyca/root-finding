```python
#finding x*sin(x)-e^x while loop
import math
def func(x):
    return x*math.sin(x)-math.e**x    
def bisection(a, b, tol):
    i=0
    Err=abs(b-a)    
    if func(a)*func(b)>0:
        print("The root can't be found!")
        print("please change a & b")        
    else:
        while Err>tol:
            c=(a+b)/2
            i=i+1
            print('{:2d} {:3.4f} {:3.4f} {:3.4f} {:3.4f} {:3.4f} {:3.4f} {:3.4f}'
                  .format(i, a, b, c, func(a), func(b), func(c), Err))            
            if func(c)==0:
                return c
            elif func(c)<0:
                b=c
            else:
                a=c
            Err=abs(b-a)
        return c

print("iter  a      b      c     f(a)   f(b)   f(c)    Err")
print("---------------------------------------------------------")
c=bisection(-1, 0, 1e-5)
if c is not None:
    print("---------------------------------------------------------")
    print(f"The Root of f(x) is ",round(c,5))
else:
    print("Roof finding failed.")
```
